```python
NY = pd.read_csv('USW00094728.csv',header=None, encoding=None, dtype=str)
dtframe = pd.DataFrame(NY)
dtframe[1] = pd.to_datetime(dtframe[1],format='%Y%m%d')
Tmin = dtframe[dtframe[2].str.contains('TMIN', regex=False,na=False)]
Tmax = dtframe[dtframe[2].str.contains('TMAX', regex=False,na=False)]
TMinAVG = Tmin[3].astype('float').mean()
TMaxAVG = Tmax[3].astype('float').mean()
YEARSAVG = (TMinAVG + TMaxAVG) / 2
print(YEARSAVG)
```

```python
MidCentury = dtframe[(dtframe[1] > '1945-01-01') & (dtframe[1] < '1955-02-01')]
TminMC = MidCentury[MidCentury[2].str.contains('TMIN', regex=False,na=False)]
TmaxMC = MidCentury[MidCentury[2].str.contains('TMAX', regex=False,na=False)]
TMinAVGMC = TminMC[3].astype('float').mean()
TMaxAVGMC = TmaxMC[3].astype('float').mean()
MCAVG = (TMinAVGMC + TMaxAVGMC) / 2
print(MCAVG)
```

```python
Anomaly = YEARSAVG - MCAVG
print(Anomaly)
```

```python
TminYEAR = Tmin[1]
TmaxYEAR = Tmax[1]
TMIN = Tmin[3].astype('float')
TMAX = Tmax[3].astype('float')

names = ['Anomaly','TMinAVG','TMaxAVG']
values = [Anomaly, TMinAVG, TMaxAVG]

plt.figure()
plt.subplot(111)
plt.bar(names, values)
plt.ylim([-25,200])
plt.title('NEW YORK ANOMALY')
plt.rcParams["figure.figsize"] = (20,10)

plt.show()
```
